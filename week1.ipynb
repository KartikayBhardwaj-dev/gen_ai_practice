##Tokens

```python

import tiktoken
enc = tiktoken.encoding_for_model("gpt-4o-mini")
text1 = "Langchain is powerfull but tricky"
tokens1 = enc.encode(text1)
print("Tokens:", tokens1)
print("Token Count:", len(tokens1))
decoded1 = enc.decode(tokens1)
print("Decoded:", decoded1)

text2 = "mai acha bacha hoon"
tokens2 = enc.encode(text2)
print("tokens", tokens2)
decoded2 = enc.decode(tokens2)
print("decoded", decoded2)
```

##Count tokens

```python
def count_token(text):
  return len(enc.encode(text))

context = "A" * 1000
print(count_token(context))
```

## Temperature vs Top-p
- These control randomness of token selection
1. Temperature - controls the sharpeness of softmax
- 0 → Almost deterministic (greedy selection)
- 0.2 – 0.5 → Controlled, factual, stable
- 0.6 – 0.9 → Creative but still coherent
- 1.0 → Quite random / diverse
- > 1.0 → Chaotic / unstable / often garbage
2. Top-p -> instead of all tokens select only smallest set whose cumulative prob = p
- ex: top_p = 0.9 -> only top 90 % probability

```python
from transformers import pipeline

pipe = pipeline("text-generation", model="distilgpt2")

prompt = "Artificial Intelligence is"

for temp in [0, 0.5, 1]:
    print(f"\nTemperature: {temp}")

    if temp == 0:
        out = pipe(
            prompt,
            max_new_tokens=30,
            do_sample=False
        )
    else:
        out = pipe(
            prompt,
            max_new_tokens=30,
            temperature=temp,
            do_sample=True
        )

    print(out[0]["generated_text"])
```

It generated same output for temp 0 but creative at temp = 0.5 but meaning less in temp 1
